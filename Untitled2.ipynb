```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np

data = input_data.read_data_sets('./tmp/data', one_hot = False)
```

```python
def one_hot_single(y):
    vector = np.zeros((y.shape[0], 9))
    
    for i in range(y.shape[0]):
        if y[i] == 6 or y[i] == 9:
            vector[i][6] = 1.
        else:
            vector[i][y[i]] = 1.
    
    return vector
```

```python
batch_size = 200

## parameters 
sigma = 1.0
weights_conv1 = tf.Variable(tf.random_normal([5, 5, 1, 32])) # First Filters
#biases_conv1 = tf.Variable(tf.random_normal([32])) 

weights_conv2 = tf.Variable(tf.random_normal([5, 5, 32, 64], stddev = sigma))
#biases_conv2 = tf.Variable(tf.random_normal([64]))

weights_fc1 = tf.Variable(tf.random_normal([7 * 7 * 64, 1024], stddev = sigma))
biases_fc1 = tf.Variable(tf.random_normal([1024], stddev = sigma))

weights_nca1 = tf.Variable(tf.random_normal([1024, 700], stddev = sigma))
biases_nca1 = tf.Variable(tf.random_normal([700], stddev = sigma))

weights_nca2 = tf.Variable(tf.random_normal([700, 300], stddev = sigma))
biases_nca2 = tf.Variable(tf.random_normal([300], stddev = sigma))

weights_nca3 = tf.Variable(tf.random_normal([300, 70], stddev = sigma))
biases_nca3 = tf.Variable(tf.random_normal([70], stddev = sigma))
# For MNIST, considering 6 and 9 as same softmax output
weights_softmax = tf.Variable(tf.random_normal([70, 9], stddev = sigma)) 
biases_softmax = tf.Variable(tf.random_normal([9], stddev = sigma))


Xa = tf.placeholder('float', [None, 28, 28, 1])
Ya = tf.placeholder('float', [None, 9])

Xb = tf.placeholder('float', [None, 28, 28, 1])
Yb = tf.placeholder('float', [None, 9])
```

```python
def cnn_nca(x, y):
    
    with tf.name_scope("Convolution_Layer_1"):
        conv1 = tf.nn.conv2d(x, weights_conv1, strides = [1, 1, 1, 1], padding = 'SAME')
        conv1 = tf.nn.relu(conv1)
        conv1 = tf.nn.max_pool(conv1, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'SAME')

    with tf.name_scope("Convolution_Layer_2"):
        conv2 = tf.nn.conv2d(conv1, weights_conv2, strides = [1, 1, 1, 1], padding = 'SAME')
        conv2 = tf.nn.relu(conv2)
        conv2 = tf.nn.max_pool(conv2, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'SAME')

    with tf.name_scope("Fully_Connected_Layer"):
        fc1 = tf.reshape(conv2, [-1, 7 * 7 * 64])
        fc1 = tf.matmul(fc1, weights_fc1) + biases_fc1
        fc1 = tf.nn.relu(fc1)

    with tf.name_scope("NCA_Subspace_Layer_1"):
        nca1 = tf.matmul(fc1, weights_nca1) + biases_nca1
        nca1 = tf.nn.relu(nca1)
    
    with tf.name_scope("NCA_Subspace_Layer_2"):
        nca2 = tf.matmul(nca1, weights_nca2) + biases_nca2
        nca2 = tf.nn.relu(nca2)
    with tf.name_scope("NCA_Subspace_Layer_3"):
        nca3 = tf.matmul(nca2, weights_nca3) + biases_nca3
        nca3 = tf.nn.relu(nca3)        
    with tf.name_scope("Output_Softmax_Layer"):
        output = tf.matmul(nca3, weights_softmax) + biases_softmax
    
    costPred = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels = y, logits = output))
    return [tf.nn.l2_normalize(nca3), output, costPred]
```

```python
def nca_objective(Xas, Yas, Xbs, Ybs, epoch_y_, same_array = []):
    
            
    ncaVectorsXas, _, _ = cnn_nca(Xas, Yas)
    ncaVectorsXbs, _, _ = cnn_nca(Xbs, Ybs)
    
    ncaVectorsXas = tf.reshape(ncaVectorsXas, [batch_size, 70])
    ncaVectorsXbs = tf.reshape(ncaVectorsXbs, [batch_size, 70])
    epsilon = tf.constant([0.00001])
    obj = tf.constant([0.])
    for i in range(batch_size):
        # Epoch_Y is [?, 1] array
        
        sameTensor = same_array[np.argmax(epoch_y_[i])]
        num = tf.exp(-1 * tf.reduce_sum(tf.squared_difference(tf.reshape(ncaVectorsXas[i], [1, 70]), \
                                                               ncaVectorsXbs) * sameTensor, axis = 1))
        num = tf.reshape(num, [batch_size, 1]) * sameTensor
        den = tf.exp(-1 * tf.reduce_sum(tf.squared_difference(tf.reshape(ncaVectorsXas[i], [1, 70]), \
                                                               ncaVectorsXbs), axis = 1))
        
        obj = obj + tf.reduce_sum(num) / (tf.reduce_sum(den) + epsilon)

    return obj
```

```python
def computeSameTensorArray(epoch_y_):
    
    same_array = [[], [], [], [], [], [], [], [], []]
    for i in range(batch_size):
        
        same_array[0].append([1. - (0 if epoch_y_[i][0] == 1 else 1)])
        same_array[1].append([1. - (0 if epoch_y_[i][1] == 1 else 1)])
        same_array[2].append([1. - (0 if epoch_y_[i][2] == 1 else 1)])
        same_array[3].append([1. - (0 if epoch_y_[i][3] == 1 else 1)])
        same_array[4].append([1. - (0 if epoch_y_[i][4] == 1 else 1)])
        same_array[5].append([1. - (0 if epoch_y_[i][5] == 1 else 1)])
        same_array[6].append([1. - (0 if epoch_y_[i][6] == 1 else 1)])
        same_array[7].append([1. - (0 if epoch_y_[i][7] == 1 else 1)])
        same_array[8].append([1. - (0 if epoch_y_[i][8] == 1 else 1)])
    return [tf.constant(same_array[0]), tf.constant(same_array[1]), tf.constant(same_array[2]), \
                  tf.constant(same_array[3]), tf.constant(same_array[4]), tf.constant(same_array[5]), \
                  tf.constant(same_array[6]), tf.constant(same_array[7]), tf.constant(same_array[8])]
```

```python
from math import pi

def rotation(Xa):
    
    angles = pi / 4 * np.random.randint(0, 8, batch_size)
    Z = tf.contrib.image.rotate(Xa, angles, interpolation = 'BILINEAR')
    
    return Z

def random_rotation(Xa):
    angles = 4 * pi * np.random.random(10000) - pi
    Z = tf.contrib.image.rotate(Xa, angles, interpolation = 'BILINEAR')
    
    return Z

data_test_images = data.test.images
data_test_images = data_test_images.reshape((data.test.num_examples, 28, 28, 1))
    
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    rotated_test_images = sess.run(random_rotation(Xa), feed_dict = {Xa : data_test_images})
    sess.close()
#rotated_test_images_ = data_test_images[0 : 1000]
rotated_test_images_ = rotated_test_images[0 : 1000]
rotated_test_labels = one_hot_single(data.test.labels)
rotated_test_labels_ = rotated_test_labels[0 : 1000]
```

```python
num_epochs = 3


nca3, output, costPred = cnn_nca(Xa, Ya)
opt = tf.train.AdamOptimizer()
sess = tf.InteractiveSession()
sess.run(tf.initialize_all_variables())
sess.run(tf.global_variables_initializer())
for i in range(num_epochs):
    for _ in range(data.train.num_examples // batch_size):

        epoch_x, epoch_y = data.train.next_batch(batch_size)
        epoch_x = sess.run(rotation(Xa), feed_dict = {Xa : epoch_x.reshape((batch_size, 28, 28, 1))})
        same_array = computeSameTensorArray(one_hot_single(epoch_y))
    
        grads_and_vars_classifying = opt.compute_gradients(costPred)
        train_opt_classifying = opt.apply_gradients(grads_and_vars_classifying)
    
        ncaObj = nca_objective(Xa, Ya, Xb, Yb, one_hot_single(epoch_y), same_array)
    
        grads_and_vars_nca = opt.compute_gradients(ncaObj)
        train_opt_nca = opt.apply_gradients(grads_and_vars_nca)
    
        _, c = sess.run([train_opt_classifying, costPred], \
                    feed_dict = {Xa : epoch_x.reshape((batch_size, 28, 28, 1)), Ya : one_hot_single(epoch_y)})
        print(c)
            
        correct = tf.equal(tf.argmax(output, 1), tf.argmax(Ya, 1))
        accuracy = tf.reduce_mean(tf.cast(correct, 'float'))
            
        z1 = accuracy.eval({Xa : rotated_test_images_, Ya : rotated_test_labels_})
        print(z1)
        print(sess.run(tf.gradients(ncaObj, weights_conv1)))
    
           
    
        _, c = sess.run([train_opt_nca, ncaObj], feed_dict = {Xa : epoch_x.reshape((batch_size, 28, 28, 1)), \
                                                Ya : one_hot_single(epoch_y), \
                                                Xb : epoch_x.reshape((batch_size, 28, 28, 1)), \
                                                Yb : one_hot_single(epoch_y)})
            print(c)
```
